```python
Lines=open("example.txt").read().strip('\n').split("\n")
fart=[]
x=[]
y=[]
for i in Lines:
   fart.extend(i.split('->')) 
for i in fart:
    x.append(int(i.split(",")[0]))
    y.append(int(i.split(",")[1]))
X=min(x)
Y=min(y)
print(X,Y)
```

```python
def norm(p):
    return p-np.array((X,0))

def print_dust(matrix):        
    for j in range(max(y)+1):
        for i in range(max(x)-min(x)+1):
            if(matrix[j,i]==1):
                print('#',end="")    
            elif(matrix[j,i]==2):
                print("+",end="")
            elif(matrix[j,i]==3):
                print("o",end="")
            else:
                print(".",end="")
        print("\n",end="")
    print('\n')

def reset_matrix():
    matrix = np.zeros((max(y)+1, max(x)-min(x)+1),dtype=int)
    
    # Add stones to matrix.
    for Line in Lines:
        line=Line.split("->")    
        for i in range( len(line)-1 ):
            p1x, p1y=norm( np.array( eval(line[i]   ) )) 
            p2x, p2y=norm( np.array( eval(line[i+1] ) )) 
            dx=int(abs(p2x-p1x))
            dy=int(abs(p2y-p1y))
            px = p1x if p1x < p2x else p2x
            py = p1y if p1y < p2y else p2y
            matrix[py:py+dy+1,px:px+dx+1] = 1
    
    # Add sand source to matrix
    sx,sy=norm(np.array( (500,0) ) )
    matrix[sy,sx]=2
    
    return matrix
    
```

```python
matrix=reset_matrix()
# Let's look at some sand.

for sand in range(29):
    sx,sy=norm(np.array( (500,1) ) )
    while(True):
        before=matrix[sy,sx]
        matrix[sy,sx]=3
        print(sy)
        #print_dust(matrix)
        #display(matrix)
        matrix[sy,sx]=before
        # If it has reached the bottom.
        if sy==max(y):
            matrix[sy,sx]=3
            break
        
        #if sy==max(y)-1:
        #    print("One from end")
        #    print(f"matrix after: {matrix[sy+1,sx]}")
            
    
        # If there is a stone now ahead:
        if matrix[sy+1,sx]>0:
            #print('Stone ahead')
            # If there is a stone ahead, and to the right and left.
            if matrix[sy+1,sx-1]>0 and matrix[sy+1,sx+1]>0:
                #print('Stone ahead and left and right')
                matrix[sy,sx]=3
                break
            # If there is a stone ahead, and to the left
            if matrix[sy+1,sx-1]>0:
                #print('Stone ahead and left')
                sy +=1
                sx +=1
                continue
            sy +=1
            sx -=1
            continue
    
        sy +=1
    print(f"sand: {sand}")
    print_dust(matrix)
```

```python
with open("input.txt") as fo:
    lines = fo.read().strip('\n').split('\n')

paths = [line.split(' -> ') for line in lines]

def initiate_chart():
    chart = {}
    minx = 500; maxx = maxy = -500

    for path in paths:
        for i in range(len(path)-1):
            xl, yl = path[i].split(',')
            xr, yr = path[i+1].split(',')
            xl = int(xl); yl = int(yl); xr = int(xr); yr = int(yr)
            maxx = max(max(xl, xr), maxx)
            minx = min(min(xl, xr), minx)
            maxy = max(max(yl, yr), maxy)
            miny = 0
            if yl == yr:
                (xl, xr) = (xl, xr) if xl <= xr else (xr, xl)
                for x in range(xl,xr+1):
                    chart[(x, yl)] = '#'
            if xl == xr:
                (yl, yr) = (yl, yr) if yl <= yr else (yr, yl)
                for y in range(yl, yr+1):
                    chart[(xl, y)] = '#'
    return chart, minx, maxx, miny, maxy


def expand_chart(chart, minx, maxx, miny, maxy):
    maxy += 2; minx -= 150; maxx += 200
    for x in range(minx, maxx):
        chart[(x, maxy)] = '#'
    return chart, minx, maxx, miny, maxy


def traverse(chart, minx, maxx, miny, maxy):
    i = 0
    while True:
        x = 500; y = 0
        while True:
            if y > maxy + 1:
                return i, chart, minx, maxx, miny, maxy
            if not chart.get((x, y+1)):
                y += 1; continue
            if not chart.get((x-1, y+1)):
                y += 1; x -= 1; continue
            if not chart.get((x+1, y+1)):
                y += 1; x += 1; continue
            chart[(x, y)] = 'o'
            if y == 0:
                i += 1
                return i, chart, minx, maxx, miny, maxy
            break
        i += 1


def show(chart, minx, maxx, miny, maxy):
    for y in range(miny, maxy + 1):
        for x in range(minx, maxx + 1):
            print(chart.get((x,y), '.'), end='')
        print('')


# Part 1
*chart_stuff, = initiate_chart()
result1, *chart_stuff = traverse(*chart_stuff)
# show(*chart_stuff)


# Part 2
*chart_stuff, = initiate_chart()
*chart_stuff, = expand_chart(*chart_stuff)
result2, *chart_stuff = traverse(*chart_stuff)
# show(*chart_stuff)

print("The result is for part 1 is:", result1)
print("The result is for part 2 is:", result2)
```
